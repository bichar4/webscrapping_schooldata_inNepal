```python
import json 
import requests
import csv
from bs4 import BeautifulSoup
import re


def opencsvFileandwriteheader(filename):
    #fields to be added in csv file 
    main_data = ['id','colz_name', 'email', 'address','district_number','phone_numbers','website','courses_offered' ]
    with open(filename,'a') as csvFile:
        writer = csv.writer(csvFile)
        writer.writerow(main_data)
    csvFile.close()
    
def requestandsave(rooturl,filename):
    response = requests.get(rooturl)
    data = json.loads(response.text)
    #accessing the first node of the data 
    for i in range(len(data['results'])):
        #link for getting specific single url of colz 
        link = 'https://edusanjal.com'+ data['results'][i]['url']
        #required for accessing the telephone data and offered courses from html pages 
        html = requests.get(link).text
        soup = BeautifulSoup(html,'html.parser')
        telephone = list(soup.find_all(href=re.compile("tel:")))
        tel_data = []
        for text in telephone:
            tel_data.append(str(text.string))
        phoneNumber= "".join(tel_data)
        courses = list((soup.find_all(href=re.compile("/course"))))
        course_data = []
        for course in courses:
            course_data.append(str(course['href'])[8:])
        all_courses = "".join(course_data)
        website = list((soup.find_all(href=re.compile("//www"))))
        url = ""
        if(website):
            url = website[0]['href']

        #these data are directly recieved from the api
        ids = data['results'][i]['id']
        title = data['results'][i]['title']
        email = data['results'][i]['email']
        address = data['results'][i]['address']
        district_number = data['results'][i]['district']
        main_data = [ids,title, email, address, district_number,phoneNumber,url,all_courses]
        with open(filename,'a') as csvFile:
            writer = csv.writer(csvFile)
            writer.writerow(main_data)
        csvFile.close()
        url = ""
    while data['next'] is not None:
        response = requests.get(data['next'])
        data = json.loads(response.text)
        for i in range(len(data['results'])):
            
            link = 'https://edusanjal.com'+ data['results'][i]['url']
            html = requests.get(link).text
            soup = BeautifulSoup(html,'html.parser')
            telephone = list(soup.find_all(href=re.compile("tel:")))
            tel_data = []
            
            for text in telephone:
                tel_data.append(str(text.string))
            phoneNumber= "".join(tel_data)
            
            courses = list((soup.find_all(href=re.compile("/course"))))
            course_data = []
            for course in courses:
                course_data.append(str(course['href'])[8:])
            all_courses = "".join(course_data)
            website = list((soup.find_all(href=re.compile("//www"))))
            url = ""
            if(website):
                url = website[0]['href']

            ids = data['results'][i]['id']
            title = data['results'][i]['title']
            email = data['results'][i]['email']
            address = data['results'][i]['address']
            district_number = data['results'][i]['district']
            main_data = [ids,title, email, address, district_number,phoneNumber,url,all_courses ]
            with open(filename,'a') as csvFile:
                writer = csv.writer(csvFile)
                writer.writerow(main_data)
            csvFile.close()
            url = ""
            print("done!!!")


rooturl = "https://edusanjal.com/api/v1/colleges/"
filename = "colz_data.csv"

school_rooturl = "https://edusanjal.com/api/v1/schools/"
school_filename = "school_data.csv"

opencsvFileandwriteheader(filename)
requestandsave(rooturl,filename)
print("finally done")
```
